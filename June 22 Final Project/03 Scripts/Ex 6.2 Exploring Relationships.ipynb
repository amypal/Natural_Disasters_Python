```python
# Create Dictionary to convert binary variables into numerical 

Convert_binary = {'OFDA Response' : {'NaN': 0, 'Yes' : 1},
                  'Appeal' : {'NaN' : 0,'No' : 0,'Yes' : 1},
                  'Declaration' : {'NaN' : 0,'No' : 0,'Yes' : 1}}
```

### Initial insights - 
There are clearly some strong correlations within the dataframe. It's impossible to tell what these may be using the above heartmap as there are no axis labels. To determine which variables I should keep I'll add some labels to this style of plot and then review my variables to create a subplot

```python
f= plt. figure(figsize =(18,18))
plt.matshow(df.corr(), fignum = f.number)
plt.xticks(range(df.shape[1]), df.columns, fontsize = 8, rotation = 55)
plt.yticks(range(df.shape[1]), df.columns, fontsize = 8)
cb = plt.colorbar()
cb.ax.tick_params(labelsize = 10)
```

# Categorical Plot
### Define categories of Total Deaths to use in categorical plots

```python
df.loc[df['Total Deaths'] < 25000, 'Deaths'] = 'Low Death Toll'
```

```python
df.loc[(df['Total Deaths'] >= 25000) & (df['Total Deaths'] < 50000), 'Deaths'] = 'Medium Death Toll'
```

```python
df.loc[(df['Total Deaths'] >= 50000) & (df['Total Deaths'] < 200000), 'Deaths'] = 'High Death Toll'
```

```python
df.loc[df['Total Deaths'] > 200000, 'Deaths'] = 'Very High Death Toll'
```

```python
df['Deaths'].value_counts()
```

```python
# Create a categorical plot uisng the Deaths category across Disaster Type 
hue_order = ['Low Death Toll','High Death Toll', 'Medium Death Toll','Very High Death Toll']


sns.set(style='ticks')
g = sns.catplot(x='Disaster Type', y='Total Deaths', hue='Deaths', data=df, height = 5, aspect = 2, hue_order = hue_order)
g = g.set(xlim=(0,6))
```

This shows some interesting patterns
- The highest death tolls occur only in the Storm, Drought and Earthquake disasters
- Earthquakes appear to be the casue of the most deaths overall. I'd like to explore this further.

```python
# Create a categorical plot uisng the Deaths category across Continents
hue_order = ['Low Death Toll','High Death Toll', 'Medium Death Toll','Very High Death Toll']


sns.set(style='ticks')
g = sns.catplot(x='Continent', y='Total Deaths', hue='Deaths', data=df, height = 5, aspect = 2, hue_order = hue_order)
g = g.set(xlim=(0,6))
```

This has also reavealed some interesting insights
- The very highest deal tolls occured only in Afric and Asia
- The majority of deaths seem to occur in Asia. This is something else I'd like to explore
