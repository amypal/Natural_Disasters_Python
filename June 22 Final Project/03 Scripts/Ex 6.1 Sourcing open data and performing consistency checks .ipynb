# 04 Check for mixed types

```python
for col in df.columns.tolist():
  weird = (df[[col]].applymap(type) != df[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df[weird]) > 0:
    print (col)
```

```python
# Addressing mixed type columns 

df['Disaster Subtype'] = df['Disaster Subtype'].astype('str')
df['Disaster Subsubtype'] = df['Disaster Subsubtype'].astype('str')
df['Event Name'] = df['Event Name'].astype('str')
df['Location'] = df['Location'].astype('str')
df['Origin'] = df['Origin'].astype('str')
df['Associated Dis'] = df['Associated Dis'].astype('str')
df['Associated Dis2'] = df['Associated Dis2'].astype('str')
df['OFDA Response'] = df['OFDA Response'].astype('str')
df['Appeal'] = df['Appeal'].astype('str')
df['Declaration'] = df['Declaration'].astype('str')
df['Dis Mag Scale'] = df['Dis Mag Scale'].astype('str')
```

# 05 Dropping columns


```python
# Will drop columns that won't aid the analysis 

df = df.drop(columns = ['Glide'])
df = df.drop(columns = ['Local Time'])
df = df.drop(columns = ['River Basin'])
df = df.drop(columns = ['Adm Level'])
df = df.drop(columns = ['Admin1 Code'])
df = df.drop(columns = ['Admin2 Code'])
df = df.drop(columns = ['Geo Locations'])
```
